```python
import re
def extract_count(cell):
    matches = [m for m in re.findall("(\d+)-(\d+)", cell)]
    if not matches:
        return 1

    first, last = matches[0]
    return int(last)-int(first)+1 
    return 1

assert extract_count("2222 Amerika gyűjtemény") == 1
assert extract_count("2209-2211 Amerika gyűjtemény") == 2211-2209+1, extract_count("2209-2211 Amerika gyűjtemény")
assert extract_count("2021.15.1 Földművelés gyűjtemény") == 1
assert extract_count("6563-6598 (F244-F279) Fényképgyűjtemény") == 6598-6563+1
assert extract_count("2021.21.1-2 Európa gyűjtemény") == 2

df['count'] = df['entry'][(df['ok']==True) & (df['drop']==False)].apply(extract_count)

df.to_excel("./torzskonyv_explode_count.xlsx", index=False)
pd.read_excel("./torzskonyv_explode_count.xlsx")
```
